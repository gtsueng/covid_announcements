```python
import os
import json
import pandas
```

```python
#### Data path
sourcepath = 'data/archive/'
trainpath = 'data/processed/'
datalist = os.listdir(sourcepath)
trainlist = os.listdir(trainpath)
```

```python
processed_list = os.listdir(trainpath)
with open(os.path.join('data','processed_list.txt'),'w') as outfile:
    for eachitem in processed_list:
        outfile.write(eachitem+'\n')
```

```python
trainingset = []

for i in range(len(trainlist)):
    with open(trainpath+trainlist[i]) as traindata:
        tmpresults = json.load(traindata)
        aboutlist = tmpresults["about"]
        for j in range(len(aboutlist)):
            trainingset.append({'id':tmpresults['_id'],'text':tmpresults['articleContent'],
                                'type':aboutlist[j]['@type'],'date':tmpresults['datePublished']})

tf = pandas.DataFrame(trainingset)
#tf.sort_values('type',ascending = True,inplace=True)
print(tf.head(n=2))
print(len(tf))
```

```python
trainlist = os.listdir(trainpath)
trainingset = []

for i in range(len(trainlist)):
    with open(trainpath+trainlist[i]) as traindata:
        tmpresults = json.load(traindata)
        aboutlist = tmpresults["about"]
        for j in range(len(aboutlist)):
            if aboutlist[j]['@type']=='PhysicalAccessInfo':
                locationlist = aboutlist[j]['locationType']
                if type(locationlist) is str:
                    changes = aboutlist[j]['changeType']
                    if type(changes) is str:
                        tmpdict = {'id':tmpresults['_id'],'type':aboutlist[j]['@type'],
                                   'date':tmpresults['datePublished'],
                                   'locationType':locationlist,
                                   'changeType':changes}
                        trainingset.append(tmpdict)
                    else:
                        for eachchange in changes:
                            tmpdict = {'id':tmpresults['_id'],'type':aboutlist[j]['@type'],
                                       'date':tmpresults['datePublished'],
                                       'locationType':locationlist,
                                       'changeType':eachchange}
                            trainingset.append(tmpdict)                           
                else:
                    for eachplace in locationlist:
                        changes = aboutlist[j]['changeType']
                        if type(changes) is str:
                            tmpdict = {'id':tmpresults['_id'],'type':aboutlist[j]['@type'],
                                       'date':tmpresults['datePublished'],
                                       'locationType':eachplace,
                                       'changeType':changes}
                            trainingset.append(tmpdict)
                        else:
                            for eachchange in changes:
                                tmpdict = {'id':tmpresults['_id'],'type':aboutlist[j]['@type'],
                                       'date':tmpresults['datePublished'],
                                       'locationType':eachplace,
                                       'changeType':eachchange}
                                trainingset.append(tmpdict)
tf = pandas.DataFrame(trainingset)

print(tf)
```

```python
location_info = tf.groupby(['locationType','changeType']).size().reset_index(name='frequency')

location_info.to_csv('location_dump.tsv',sep='\t',header=True)
```

```python
trainlist = os.listdir(trainpath)
trainingset = []

for i in range(len(trainlist)):
    with open(trainpath+trainlist[i]) as traindata:
        tmpresults = json.load(traindata)
        aboutlist = tmpresults["about"]
        for j in range(len(aboutlist)):
            if aboutlist[j]['@type']=='PoliciesAndGuidelines':
                policylist = aboutlist[j]['policyType']
                if type(policylist) is str:
                    subjects = aboutlist[j]['subject']
                    if type(subjects) is str:
                        tmpdict = {'id':tmpresults['_id'],'type':aboutlist[j]['@type'],
                                   'date':tmpresults['datePublished'],
                                   'policyType':policylist,
                                   'subject':subjects}
                        trainingset.append(tmpdict)
                    else:
                        for eachsubject in subjects:
                            tmpdict = {'id':tmpresults['_id'],'type':aboutlist[j]['@type'],
                                       'date':tmpresults['datePublished'],
                                       'policyType':policylist,
                                       'subject':eachsubject}
                            trainingset.append(tmpdict)                           
                else:
                    for eachpolicy in policylist:
                        subjects = aboutlist[j]['subject']
                        if type(subjects) is str:
                            tmpdict = {'id':tmpresults['_id'],'type':aboutlist[j]['@type'],
                                       'date':tmpresults['datePublished'],
                                       'policyType':eachpolicy,
                                       'subject':subjects}
                            trainingset.append(tmpdict)
                        else:
                            for eachsubject in subjects:
                                tmpdict = {'id':tmpresults['_id'],'type':aboutlist[j]['@type'],
                                       'date':tmpresults['datePublished'],
                                       'policyType':eachpolicy,
                                       'subject':eachsubject}
                                trainingset.append(tmpdict)
tf = pandas.DataFrame(trainingset)

#print(tf['policyType'].unique().tolist())
#print(tf.loc[tf['subject']=='physical'])


#policies = tf.groupby(['policyType','subject']).size().reset_index(name='frequency')

#policies.to_csv('policy.tsv', sep='\t', header=True)

clean = tf.drop_duplicates(subset=(['id','subject']), keep="first")
policies_clean = tf.groupby(['subject']).size().reset_index(name='frequency')
policies_clean.to_csv('policy_clean.tsv', sep='\t', header=True)
```

```python
trainlist = os.listdir(trainpath)
trainingset = []

for i in range(len(trainlist)):
    with open(trainpath+trainlist[i]) as traindata:
        tmpresults = json.load(traindata)
        aboutlist = tmpresults["about"]
        for j in range(len(aboutlist)):
            if aboutlist[j]['@type']=="DiseaseNotice":
                try:
                    medcon = aboutlist[j]['medicalCondition']["name"]
                except:
                    medcon = "None"
                try:
                    infagent = aboutlist[j]['infectiousAgent']["name"].lower()
                    infdis = aboutlist[j]['infectiousDisease']["name"].lower()
                except:
                    infagent = "None"
                    infdis = "None"
                notices = aboutlist[j]['noticeType']
                if type(notices) is str:
                    tmpdict = {'id':tmpresults['_id'],'type':aboutlist[j]['@type'],
                               'date':tmpresults['datePublished'],
                               'infectiousAgent':infagent,
                               'infectiousDisease':infdis,
                               'medicalCondition':medcon,
                               'noticeType':notices}
                    trainingset.append(tmpdict)
                else:
                    for eachnotice in notices:
                        tmpdict = {'id':tmpresults['_id'],'type':aboutlist[j]['@type'],
                               'date':tmpresults['datePublished'],
                               'infectiousAgent':infagent,
                               'infectiousDisease':infdis,
                               'medicalCondition':medcon,
                               'noticeType':eachnotice}
                        trainingset.append(tmpdict)

tf = pandas.DataFrame(trainingset)

print(tf)
```

```python
disease_notices = tf.groupby(['id','infectiousDisease']).size().reset_index(name='noticetypecount')
disease_stats = disease_notices.groupby('infectiousDisease').size().reset_index(name='frequency')
print(disease_stats)

condition_notices = tf.groupby(['id','medicalCondition']).size().reset_index(name='noticetypecount')
condition_stats = condition_notices.groupby('medicalCondition').size().reset_index(name='frequency')
print(condition_stats)

print(tf.loc[tf['medicalCondition']=='mental disorder'])
```
